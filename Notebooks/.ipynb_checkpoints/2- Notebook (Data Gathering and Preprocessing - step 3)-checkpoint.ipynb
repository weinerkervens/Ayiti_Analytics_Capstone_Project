```python
import pandas as pd
import numpy as np
```

```python
df_1 = pd.read_csv(r'../output/Notebook 1/number-of-facilities-by-department_ht.csv')
df_2 = pd.read_csv(r'../output/Notebook 1/covid_haiti.csv')
display(df_1)
display(df_2)
```

```python
df = df_1.merge(df_2, on='Département', how = 'right')
df
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="my_app")
```

```python
air = pd.read_csv(r'../output/Notebook 0/ht_airports.csv')
air.head()
```

```python
air['Département'] = (air.latitude_deg.astype('str') + ' , ' + air.longitude_deg.astype('str')).apply(lambda x: geolocator.reverse(x).raw['address']['state'])

# Create a list of prefix to rename every single element in the 'Département' column
list__ = ["Département de l'", 'Département du ', 'Département des ', 'Département de la ']

for replacer in list__:
    air['Département'] = air['Département'].str.replace(replacer, '')
    
# Replace value to reach our naming convention for departments
air['Département'] = air['Département'].replace('Grande-Anse',"Grande'Anse")

air
```

```python
cols_list = air.columns.tolist()

air = air[[cols_list[0], cols_list[-1], cols_list[2], cols_list[1], cols_list[3], cols_list[4]]].rename(columns={'name':'airport'})
air
```

```python
air.to_csv(r'../output/Notebook 2/airports.csv', index = False)
```

```python
airports = air.groupby(['Département']).count().reset_index()[['Département', 'airport']]
airports
```

```python
df_complete = df.merge(airports, on='Département', how = 'outer').fillna(0)
df_complete
```

```python
df_complete.to_csv(r'../output/Notebook 2/complete_dataframe.csv', index = False)
```
